```python
X
```

```python
sc = StandardScaler()
x = sc.fit_transform(X)
x
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x, Y, test_size=.20,random_state=42)
print("Train_size",X_train.shape)
print("Test_size",X_test.shape)
```

```python
lm = LinearRegression()
lm.fit(X_train,Y_train)
# regression coefficients
print('Coefficients: ', lm.coef_)
```

```python
# prediction train
pred_train = lm.predict(X_train)
pred_train
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(r2_score(Y_train,pred_train))
```

```python
# prediction test
pred_test = lm.predict(X_test)
```

```python
print(r2_score(Y_test,pred_test))
```

```python
mse = mean_squared_error(Y_test,pred_test)
mse
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(Y_test,pred_test)
mse = mean_squared_error(Y_test,pred_test)
rmse = mean_squared_error(Y_test,pred_test, squared=False)
r2 = r2_score(Y_test,pred_test)
```

```python
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
```

```python
res = pd.DataFrame({"Actual_price":Y_test, "prediction":pred_test})
res
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
num_folds = 5
# Create a k-fold cross-validation object
kf = KFold(n_splits=num_folds)
# 100 we created 5 split in data - each split - 20%
```

```python
# Perform cross-validation
scores = cross_val_score(lm,x,Y, cv=kf)

print("Cross-Validation Scores:", scores)
print("Average Score:", scores.mean())
```
